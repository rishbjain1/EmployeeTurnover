```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
import numpy as np

# Define categorical and numerical columns
categorical_cols_onehot = ['sales']  # Columns to one-hot encode
categorical_cols_label = ['salary']  # Columns to label encode
numerical_cols = ['satisfaction_level', 'number_project', 
                  'average_montly_hours', 'time_spend_company', 'promotion_last_5years']
non_scaled_cols = ['last_evaluation', 'Work_accident', 'left']  # Columns to keep as-is

# Label encode the 'salary' column
label_encoder = LabelEncoder()
df['salary'] = label_encoder.fit_transform(df['salary'])

# Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Preprocessing for categorical data (one-hot encoding)
categorical_transformer_onehot = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop='first'))
])

# Combine preprocessors in a column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat_onehot', categorical_transformer_onehot, categorical_cols_onehot)
    ])

# Apply preprocessing pipeline
X = preprocessor.fit_transform(df)

# Extract one-hot encoded column names
onehot_columns = preprocessor.named_transformers_['cat_onehot'].named_steps['onehot'].get_feature_names_out(categorical_cols_onehot)

# Combine numerical, one-hot encoded, and label-encoded column names
final_columns = numerical_cols + list(onehot_columns) + categorical_cols_label

# Add the label-encoded 'salary' column back to the transformed data
X = np.hstack([X, df[['salary']].values])

# Extract the non-scaled columns
non_scaled_data = df[non_scaled_cols].reset_index(drop=True)

# Combine the processed data with the non-scaled columns
X_final = np.hstack([X, non_scaled_data.values])

# Update the final column names
final_columns += non_scaled_cols

# Convert the transformed data back to a DataFrame
processed_df = pd.DataFrame(X_final, columns=final_columns)

# Inspect the processed DataFrame
print(processed_df.head())
```

```python
# Split the data into features and target variable
X = processed_df.drop(columns=['left'])
y = processed_df['left']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Check the class distribution after applying SMOTE
print(y_train_balanced.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Create a logistic regression model
model = LogisticRegression()

# Perform five-fold cross-validation
scores = cross_val_score(model, X_train_balanced, y_train_balanced, cv=5)

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())

# Fit the logistic regression model
model.fit(X_train_balanced, y_train_balanced)
```

```python
from sklearn.metrics import classification_report

# Make predictions on the test set
y_pred = model.predict(X_test)

# Generate the classification report
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Generate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix as a nice big table
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

To decide whether to focus on **recall** or **precision**, consider the business context of the project:

### Project Context:
The goal is to predict employee turnover (`left` column). The HR department likely wants to identify employees at risk of leaving to take preventive action. Missing employees who are likely to leave (false negatives) could result in losing valuable talent, which is costly for the company.

### Metric to Use:
- **Recall** should be prioritized in this case because:
  - It measures the ability to correctly identify all employees who are at risk of leaving.
  - A high recall ensures that fewer employees who are likely to leave are missed (minimizing false negatives).

### Why Not Precision?
- Precision focuses on minimizing false positives (employees predicted to leave but actually stay). While important, false positives are less critical in this context because intervening with employees who are not at risk is less costly than missing those who are.

### Conclusion:
From the confusion matrix, **recall** is the key metric to focus on for this project.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Perform five-fold cross-validation and predict probabilities on test data
y_prob = cross_val_predict(rf_classifier, X_test, y_test, cv=5, method='predict_proba')

# Extract the probability of the positive class (employee leaving)
y_prob_leave = y_prob[:, 1]

# Predict the class labels using cross-validation
y_pred = cross_val_predict(rf_classifier, X_test, y_test, cv=5, method='predict')

# Create a DataFrame with actual labels, predicted labels, and predicted probabilities
results_df = pd.DataFrame({
    'Actual Label': y_test,
    'Predicted Label': y_pred,
    'Predicted Probability (Leaving)': y_prob_leave
})

# Display the first few rows of the DataFrame
print(results_df)
```

```python
# Define a function to categorize employees into zones based on probability scores
def categorize_zone(probability):
    if probability < 0.2:
        return 'Safe Zone (Green)'
    elif 0.2 <= probability < 0.6:
        return 'Low-Risk Zone (Yellow)'
    elif 0.6 <= probability < 0.9:
        return 'Medium-Risk Zone (Orange)'
    else:
        return 'High-Risk Zone (Red)'

# Apply the function to create a new column for zones
results_df['Risk Zone'] = results_df['Predicted Probability (Leaving)'].apply(categorize_zone)

# Display the updated DataFrame
print(results_df)
```

### Retention Strategies:
- **Safe Zone (Green)**:
  - Employees are satisfied and unlikely to leave.
  - Continue current engagement strategies and monitor periodically.

- **Low-Risk Zone (Yellow)**:
  - Employees show some signs of dissatisfaction.
  - Conduct regular check-ins and provide opportunities for growth or feedback.

- **Medium-Risk Zone (Orange)**:
  - Employees are moderately likely to leave.
  - Identify specific concerns through surveys or one-on-one meetings and address them proactively.

- **High-Risk Zone (Red)**:
  - Employees are highly likely to leave.
  - Take immediate action, such as offering promotions, salary adjustments, or addressing workplace issues to retain them.
